```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample geoscience-like dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=12, freq='M'),
    'Rainfall_mm': [120, 80, 60, 30, 10, 5, 15, 40, 80, 100, 110, 130],
    'Temperature_C': [28, 30, 32, 35, 38, 40, 39, 36, 33, 30, 29, 27]
}

df = pd.DataFrame(data)

# Preview the dataset
print(df.head())

# Plot Rainfall and Temperature over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Rainfall_mm'], label='Rainfall (mm)', marker='o')
plt.plot(df['Date'], df['Temperature_C'], label='Temperature (C)', color='orange', marker='s')

plt.title('Rainfall and Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
